### Let's do Regression 

Now, let's perform a number of simple regression experiments between the chosen independent variables and the dependent variable (price). You'll do this in a loop and in every iteration, you should pick one of the independent variables. Perform the following steps:

```python
# Your code here
import statsmodels.api as sm
from statsmodels.formula.api import ols

print("Boston Housing DataSet - Regression Analysis and Diagnostics for formula: medv~crim")

f = 'medv~crim'
model = ols(formula=f, data=data).fit()
model.summary()
```

```python
# Your code here
#Decided to make a function that will take in the model and return the y_pred line dataframe

def get_y_pred(ols_model, df, column_name):
    min_max = pd.DataFrame([df[column_name].min(), df[column_name].max()], columns=[column_name])
    min_max['y_pred'] = ols_model.predict(min_max)
    return min_max
```

```python
y_pred = get_y_pred(model, data, 'crim')

plt.scatter(data['crim'], data['medv'])
plt.plot(y_pred['crim'], y_pred['y_pred'],color='red')
```

```python
#RESIDUALS PLOTS
residuals_plot = plt.figure(figsize=(15,8))
sm.graphics.plot_regress_exog(model, "crim", fig=residuals_plot)
```

```python
#QQ PLOTS
import scipy.stats as stats
residuals = model.resid
fig = sm.graphics.qqplot(residuals, dist=stats.norm,line='45',fit=True)
fig.show()
```
